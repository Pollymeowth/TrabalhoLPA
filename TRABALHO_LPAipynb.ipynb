#Atividade 1

```python
print("                           Bem vindo a loja da Pollyana Roberta de Sousa!")
#gostaria que o texto aparecesse mais centralizado.
valor=float(input("\nQual o valor do produto? R$"))
qtdProduto=int(input("Quantas unidades desse produto você deseja adquirir(a partir de 10 unidades)?"))#Nao sabia
#se era necessario incluir uma linha de código que o programa lesse (porque não era uma exigencia), mas achei importante
#tambem informar ao usuario, já que havia essa observação no material.
preco=valor*qtdProduto #preco do item sem desconto

#No material informa que deve ser um valor maior que 10. Se o usuario digitar qualquer valor menor que 10(ainda que ele
#tenha sido informado), já encerra o programa sem realizar calculo.
if qtdProduto <=10:
    print("Valor inválido.")

#Se quantidade for menor que 200 o desconto será de 0% (não há desconto no valor do produto),por isso já pode
#printar isso e finalizar o programa.
elif qtdProduto <200:
    print('Essa quantidade não dá direito à desconto.\nO valor SEM desconto é de R${}.' .format(valor*qtdProduto))

# Se quantidade for igual ou maior que 200 e menor que 1000 o desconto será de 5%;
elif(qtdProduto>=200) and (qtdProduto <1000):
     desconto=(valor*qtdProduto)*0.05
     print("Você ganhou 5% de desconto!\nO valor SEM desconto é de R${}.".format(valor * qtdProduto))
     print("O valor COM desconto é de R${}.".format(preco - desconto))

#Se quantidade for igual ou maior que 1000 e menor que 2000 o desconto será de 10%;
elif qtdProduto>=1000 and qtdProduto<2000:
     desconto=(valor*qtdProduto) * 0.1
     print("Você ganhou 10% de desconto!\nO valor SEM desconto é de R${}.".format(valor * qtdProduto))
     print("O valor COM desconto é de R${}.".format(preco - desconto))

#Todos os outros valores já estão contemplados, então se for qualquer outro valor além dos contemplados
#(no caso aqui, se quantidade for igual ou maior que 2000 o desconto será de 15%;), ele fara esse calculo:
else:
    desconto=(valor*qtdProduto) * 0.15
    print("Você ganhou 15% de desconto!\nO valor SEM desconto é de R${}.".format(valor*qtdProduto))
    print("O valor COM desconto é de R${}.".format(preco-desconto))
print("\nObrigada pela preferência!")
#Nao sei havia alguma restricao para adicionar elementos, mas gostaria de fazer codigos que tambem valorizassem
#a experiencia do usuario, por isso informei o valor do desconto e tambem coloquei uma saudacao final.
#Obrigada pela correcao! :)
```

```python
#-----------INICIO VARIÁVEIS GLOBAIS-----------

lista_colaboradores= []
id_global=0 #fazer a contagem de colaboradores cadastrados

#-----------FIM VARIÁVEIS GLOBAIS-----------

#-----------Inicio cadastrar_colaborador-----------
def cadastrar_colaborador(id):
  print("**Bem vindo ao menu de cadastro de colaboradores**")
  # print("Id do colaborador: {}". format(id))
  nome =input("Qual o nome do colaborador?")
  setor=input("Em qual setor o colaborador trabalha?")
  salario=int(input("Qual o salário(R$) do colaborador?"))
  dic_colaborador = {"ID"     :id,      #cria um dicionario que recebe o cadastro do colaborador com as chaves especificadas e todas essas variaveis entram na parte dos dados
                   "Nome"   : nome,
                   "Setor"  : setor,
                   "Salário(R$)":salario}
  lista_colaboradores.append(dic_colaborador.copy()) #cria uma cópia do dicionario dentro da lista de colaboradores

#-----------Fim cadastrar_colaborador-----------

#-----------Inicio consultar_colaborador-----------
def consultar_colaborador():
  print("**Bem vindo ao menu de consulta de colaboradores**")
  while True: #cria um loop infinito
    opcaoConsulta=input("\nEscolha a opção desejada: \n"+
                      "1- Consultar Todos \n"+
                      "2- Consultar por ID \n"+
                      "3- Consultar por Setor \n"+
                      "4- Retornar ao Menu Principal\n"+
                      ">>")
    if opcaoConsulta=="1":
      print("Você escolheu consultar todos os colabores.\n")
      for colaborador in lista_colaboradores: #varrer toda a lista de colaboradores
        print("-"*20)
        for key,value in colaborador.items(): #varrer todos os itens chave/valor do dicionario e printar todos
          print("{}: {}".format(key,value))

    elif opcaoConsulta=="2":
      print("Você escolheu consultar por ID.\n")
      idDesejado=int(input("Digite qual ID você deseja consultar: "))
      for colaborador in lista_colaboradores:
        if colaborador["ID"]== idDesejado: #se o valor da chave ID do dicionario for igual ao valor desejado, entra no if
          print("-"*20)
          for key,value in colaborador.items(): #varrer a chave ID e o respectivo dado do dicionario e printa
            print("{}: {}".format(key,value))

    elif opcaoConsulta=="3":
      print("Você escolheu consultar por setor.\n")
      setorDesejado=input("Digite qual setor você deseja consultar: ")
      for colaborador in lista_colaboradores:
        if colaborador["Setor"]== setorDesejado: #o valor da chave setor do dicionario for igual ao valor desejado, entra no if
          print("-"*20)
          for key,value in colaborador.items(): #varrer a chave Setor e respectivo dado do dicionario e printa
            print("{}: {}".format(key,value))

    elif opcaoConsulta=="4":
      print("Você escolheu voltar ao menu principal.\n")
      return #sai da funcao e retorna pro menu principal

    else:
      print("Opção inválida! Tente novamente.\n")
      continue #volta para o inicio do loop


#-----------Fim consultar_colaborador-----------

#-----------Inicio remover_colaborador-----------
def remover_colaborador():
  print("Bem vindo ao menu de remoção de colaboradores!")
  removeColaborador=int(input("Digite o ID do funcionário que deseja remover: "))
  for colaborador in lista_colaboradores:
    if colaborador["ID"]== removeColaborador:# o valor da chave ID do dicionario for igual ao valor desejado, entra no if
      print("-"*20)
      lista_colaboradores.remove(colaborador) #remove colaborador a partir do ID
      print("Colaborador removido com sucesso!")

#-----------Fim remover_colaborador-----------

#-----------Inicio Main -----------
print("Bem vindos ao Controle de Colaboradores de Pollyana Roberta de Sousa")
while True:#cria loop infinito
  opcaoPrincipal=input("\nEscolha a opção desejada: \n"+
                     "1- Cadastrar Colaborador\n"+
                     "2- Consultar Colaborador\n"+
                     "3- Remover Colaborador\n"+
                     "4- Encerrar Programa\n"+
                     ">>")
  if opcaoPrincipal=="1":
    id_global+=1 #faz a contagem na variavel id_global
    cadastrar_colaborador(id_global) #chama a funcao para cadastrar colaborador
  elif opcaoPrincipal=="2":
    consultar_colaborador() #chama a funcao para consulta de colaboradores
  elif opcaoPrincipal=="3":
    remover_colaborador() #chama a funcao de remocao de colaboradores
  elif opcaoPrincipal=="4":
    print("Encerrando o programa...")
    break #encerra o loop e o programa acaba
  else:
    print("Opção inválida! Tente novamente.")
    continue #volta para o inicio do loop


#-----------Fim Main -----------
```
